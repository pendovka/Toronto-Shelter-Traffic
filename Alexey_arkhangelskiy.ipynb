## Q3

```python
# Let us examine if there are any NaN in the columns we need to use:

print(admissions_img['discharge_date'].isnull().sum())
print(admissions_img['discharge_time'].isnull().sum())
print(admissions_img['admission_date'].isnull().sum())
print(admissions_img['admission_time'].isnull().sum())

# Some of the 'discharge_time' entries are NaN. since they constitute a big portion of our data, we cannot drop them.
# Let us instead convert those NaN to '00:00:00'
admissions_img['discharge_time'].fillna('00:00:00', inplace=True)
```

```python
# The time format of disharge_time and admission_time varies. Need to adress that and format time as HH:MM

def time_format(time_str):
    parts = time_str.split(':')
    return parts[0] + ':' + parts[1]

admissions_img['admission_time'] = admissions_img['admission_time'].apply(time_format)
admissions_img['discharge_time'] = admissions_img['discharge_time'].apply(time_format)

admissions_img['admission_datetime'] = pd.to_datetime(admissions_img['admission_date'] + ' ' + admissions_img['admission_time'], errors='coerce')
admissions_img['discharge_datetime'] = pd.to_datetime(admissions_img['discharge_date'] + ' ' + admissions_img['discharge_time'], errors='coerce')

admissions_img['length_of_stay'] = admissions_img['discharge_datetime'] - admissions_img['admission_datetime']

# The task requires length_of_stay to be measured in days
admissions_img['length_of_stay'] = admissions_img['length_of_stay'].dt.days

# Mean length of stay for each department
admissions_img.groupby('department').length_of_stay.mean() 
```

```python
img_mapper = imaging.test_name.value_counts().reset_index()
img_mapper.columns = ['test_name', 'frequency']
img_mapper.head()
```

# b)

```python
img_mapper['test_name'] = img_mapper['test_name'].str.lower() #bring test_name to standard format

def classify_test(test_name):
    if "us" in test_name in test_name:
        return "US"
    elif "ct" in test_name:
        return "CT"
    else:
        return "Others"

img_mapper['mapped_test'] = img_mapper['test_name'].apply(classify_test)

img_mapper.head(10)
```

# c)

```python
def classify_bodypart(test_name):
    if "neck" in test_name or "head" in test_name:
        return "Neck/Head"
    elif "abdomen" in test_name or "pelvis" in test_name:
        return "Abdomen/pelvis"
    else:
        return "Others"

img_mapper['mapped_bodypart'] = img_mapper['test_name'].apply(classify_bodypart)

print(img_mapper.head(10))
```
